```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data['Price'], kde=True)
plt.title('Distribution of Product Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Box plot for Price by Skin Type
sns.boxplot(x='Combination', y='Price', data=data)
plt.title('Price Distribution by Skin Type')
plt.xlabel('Skin Type')
plt.ylabel('Price')
plt.show()
```

```python
# Average price per brand
avg_price_per_brand = data.groupby('Brand')['Price'].mean().reset_index()
sns.barplot(x='Brand', y='Price', data=avg_price_per_brand)
plt.title('Average Price per Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Scatter plot for Price vs Rank
sns.scatterplot(x='Price', y='Rank', data=data)
plt.title('Price vs Rank')
plt.xlabel('Price')
plt.ylabel('Rank')
plt.show()
```

```python
# Pie chart for skin type distribution
skin_type_counts = data[['Combination', 'Dry', 'Normal', 'Oily', 'Sensitive']].sum()
plt.pie(skin_type_counts, labels=skin_type_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Products by Skin Type')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_1samp

population_mean = 100
t_stat, p_value = ttest_1samp(data['Price'], population_mean)

print("T-statistic:", t_stat)
print("P-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average price is significantly different from $100.")
else:
    print("Fail to reject the null hypothesis: The average price is not significantly different from $100.")
```

```python
import pandas as pd
from scipy.stats import ttest_ind

combination_prices = data[data['Combination'] == 1]['Price']
dry_prices = data[data['Dry'] == 1]['Price']
t_stat, p_value = ttest_ind(combination_prices, dry_prices)
print("T-statistic:", t_stat)
print("P-value:", p_value)
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The average prices are significantly different.")
else:
    print("Fail to reject the null hypothesis: The average prices are not significantly different.")
```
